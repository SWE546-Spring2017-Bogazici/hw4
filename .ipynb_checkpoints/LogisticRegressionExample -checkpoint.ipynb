# Maximum Likelihood for a coin toss


$
x_i \in \{H, T\}
$


$
p(x_i = T | \pi) = \pi
$
$
p(x_i = H | \pi) = (1-\pi)
$

$$
[f(x_i)] = \left\{\begin{array}{cc} 1 & if f(x_i) \text{is true} \\ 0 & \text{ow} \end{array}\right. 
$$


$$
p(x_i | \pi) = \pi^{[x_i = H]} (1-\pi)^{[x_i = T]}
$$



What is the probability of observing $X = H T H H T T T H T$?

$
p(X|\pi) = p(x_1 = H | \pi) p(x_2 = T | \pi) p(x_3 = H | \pi) \cdots p(x_9 = T | \pi)
$

$$
p(X|\pi) = \prod_{i=1}^9 p(x_i|\pi) = \prod_{i=1}^9\pi^{[x_i = H]} (1-\pi)^{[x_i = T]}
$$

## The log-likelihood function

\begin{eqnarray}
{\mathcal L}(\pi) & = & \log p(X|\pi) =  \log \prod_{i=1}^9\pi^{[x_i = H]} (1-\pi)^{[x_i = T]} \\
& = &  \sum_{i=1}^9 {[x_i = H]} \log \pi +  {[x_i = T]} \log (1-\pi) \\
& = &  \log \pi \sum_{i=1}^9 {[x_i = H]}  +  \log (1-\pi) \sum_{i=1}^9 {[x_i = T]}  \\
C_H & = & \sum_{i=1}^9 {[x_i = H]} \\
C_T & = & \sum_{i=1}^9 {[x_i = T]}
\end{eqnarray}
$$
{\mathcal L}(\pi)  =  C_H \log \pi   +  C_T \log (1-\pi)
$$

$$
\frac{\partial}{\partial \pi}{\mathcal L}(\pi)  =  C_H \frac{1}{\pi}   -  C_T \frac{1}{1-\pi} = 0
$$

$$
C_H    =  (C_T + C_H) \pi
$$

$$
\pi^*  =  \frac{C_H}{C_T + C_H}   
$$


```python
# %load /Users/cemgil/src/ipynb/notes/template.py
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
import pandas as pd

C_H = 50
C_T = 50

def ll(p):
    return C_H*np.log(p) +  C_T*np.log(1-p)

def dll(p):
     return C_H/p -  C_T/(1-p) 
        

p = np.linspace(0,1,100)

plt.figure(figsize=(5,5))

plt.plot(p, ll(p))

plt.show()

```

```python

def sigmoid(x):
    return 1/(np.exp(-x)+1)


N = 20

x = np.random.choice(range(255),size=N)


w1 = -0.04
w0 = 1.9

p = sigmoid(w1*x + w0)

plt.plot(x, p, 'o')

c = [np.random.rand()<p[i] for i in range(N)]
```

$\mathcal{L}(w_1, w_0) =  \log (1 - \sigma(w_1 \times 44 + w_0)) + \log (1 - \sigma(w_1 \times 101 + w_0)) + \log (1 - \sigma(w_1 \times 241 + w_0))+ \log \sigma(w_1 \times 110 + w_0) \dots$

$$
\mathcal{L}(w_1, w_0) =   \sum_{i=1}^N [c_i = 0]\log (1 - \sigma(w_1 x_i + w_0)) +  \sum_{i=1}^N [c_i = 1] \log \sigma(w_1 x_i + w_0) 
$$


```python
v_0, v_1 = np.meshgrid(np.linspace(-2,10, 400), np.linspace(-1,1, 200))

def ll(w0, w1):
    LL = np.zeros_like(w0)
    
    for i in range(len(x)):
        LL += np.log(1-sigmoid(w1*x[i] + w0)) if c[i]==0 else np.log(sigmoid(w1*x[i] + w0))
    
    return LL


X = np.c_[x, np.ones_like(x)]
y = np.c_[c]

def dll(w0, w1):
    pr = sigmoid(X.dot(np.mat([w1, w0],).T))
    return X.T*(y-pr)

LL = ll(v_0, v_1)

plt.imshow(np.exp(LL), interpolation='nearest')
plt.show()
```
